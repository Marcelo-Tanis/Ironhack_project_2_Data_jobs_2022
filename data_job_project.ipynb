```python
#creating a cerntralized parametrization for SQL connection
class Params:
    def __init__(self,db_server = 'pymysql',user = 'root',db_port = '3306', ip = 'localhost',db_name='IH_project2'):
        self.db_server = db_server
        self.user = user
        self.db_port = db_port
        self.password = getpass.getpass()
        self.ip = ip 
        self.db_name = db_name
```

### creating the function for SQL connection verification according to the class created above

def connection_database(params):
    try:
        engine = db.create_engine(f'mysql+{params.db_server}://{params.user}:{params.password}@{params.ip}:{params.db_port}/{params.db_name}?charset=utf8')
        conn = engine.connect()
        return conn
    except:
        print(f'Something went wron withe the connection')

```python
#create function to extract the url file and to check connection
def extract(url):
    try: 
        get_html = requests.get(url)
        if  get_html.status_code == 200:
            get_html.encoding
            html = get_html.content
            soup = BeautifulSoup(html)
            logger = logging.getLogger('get_url')
            logging.info(f'get_url successfull')
            return soup
        else:
            print(f'url {url} unavailable')
            logging.getLogger('url_connection')
            logging.info(f'url {url} unavailable')
    except:
        logger = logging.getlogger('get_url')
        logging.info(f'something went wrong in get_url')
                                     
```

```python
#creating variable soup from the function 
soup = extract(url)
```

```python
#findind the jobs' links and creating the related list using bs4 library
soup = extract(url)
jobs = soup.find_all('a',attrs={'class':"jcs-JobTitle css-jspxzf eu4oa1w0"})
job_links= ['https://br.indeed.com/viewjob'+ x.get('href')[7:] for x in jobs]
job_links_list=[link for link in job_links]
job_links_list
```

```python
#Transform step (handling, cleaning) applying the 'fail fast' data pipeline principle

job_title_count = df['job_title'].value_counts() 
df
```

```python
df['Job_title'] = df['job_title'].apply(lambda x: x if x in job_title_count.head(20) else np.nan)
df
```

```python
df.drop(['salary','salary_currency','employee_residence'], axis=1,inplace=True)
df
```

```python
df.drop(columns=['job_title'],inplace=True)
df
```

```python
colnames = ['Job_title','experience_level','employment_type','salary_in_usd','remote_ratio','company_location','company_size','work_year'] 
df = df.loc[:,colnames]
df
```

```python
df['work_year']= df['work_year'].apply(lambda x: str(x) [:4])
df
```

```python
df['monthly_salary_usd'] = df['salary_in_usd']/12
df
```

```python

df['company_size']=df['company_size'].str.replace('S','Small').replace('M','Medium').replace('L','Large')
df = df.dropna()
df
```

```python
df_data_jobs = df
df
```
